```python
import pandas as pd
census_df=pd.read_csv("census.csv")
census_df
```

```python
x=census_df.groupby('STNAME').count()['CTYNAME']
ans=x.idxmax()
print(ans)
```

```python
top_counties_df=census_df.sort_values(by=['STNAME','CENSUS2010POP'],ascending=False).groupby('STNAME').head(3)
ans=top_counties_df.groupby('STNAME').sum().sort_values(by='CENSUS2010POP',ascending=False).head(3).index.tolist()
print(ans)
```

```python
ans=census_df.groupby('STNAME').sum().sort_values(by='CENSUS2010POP',ascending=False).head(3).index.tolist()
print(ans)
```
