# Feature Scaling-to do standardizatisation,preprocessing is done


```python
#scaling using standardization or normalisation
#standardization mean=0 ,sd=1
#normalization min to max value 0,1 respectively
#when difference in units of data, feature scaling and pre processing is performed, for obtaining more accuracy
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
ss.fit(X) #training
```

```python
X_ss=ss.transform(X)
```

```python
X_ss
```

```python
import pandas as pd
X_std=pd.DataFrame(X_ss,columns=['sepal_length','sepal_width','petal_length','petal_width'])
```

```python
X_std
```

```python
X_std.describe().round(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms= MinMaxScaler()
```

```python
mms.fit(X)
```

```python
X_mms = mms.transform(X)
```

```python
X_mms
```

```python
X_nor = pd.DataFrame(X_mms,columns=['sepal_length','sepal_width','petal_length','petal_width'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_std,Y,test_size=0.3,random_state=1)
```

```python
X_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
kmodel = KNeighborsClassifier(n_neighbors=3)
```

```python
kmodel.fit(X_train,y_train)
```

```python
y_predict = kmodel.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test,y_predict)
```

```python
acc
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
cm
```
